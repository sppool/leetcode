### Find the Difference

```python
class Solution:
    def findTheDifference(self, s, t): # 98.95%
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        num = 0
        for x in t:
            num += ord(x)
        for x in s:
            num -= ord(x)
        return chr(num)
        
    def findTheDifference1(self, s, t): # 82.95%
        dic = {}
        for x in s+t:
            if x not in dic:
                dic[x] = 1
            else:
                dic[x] += 1
        for key, n in dic.items():
            if n % 2 == 1:
                return key
        
    def findTheDifference2(self, s, t): # 38.06%
        dic = {}
        for x in s:
            if x not in dic:
                dic[x] = 1
            else:
                dic[x] += 1
        for x in t:
            if x in dic:
                dic[x] -= 1
                if dic[x] == -1:
                    return x
            else:
                return x
```

```python
s = "abcd"
t = "abcde"
ans = Solution()
ans.findTheDifference(s, t)
```
